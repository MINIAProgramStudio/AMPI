```python
import numpy as np
import pyswarms as ps
from pyswarms.utils.functions import single_obj as fx
```

$f\left(X\right)=\sum_{i=1}^{n} x_{i}^{2}$, $X=\left(x_1;x_2;x_3;...;x_n\right)$.

```python
def funcSphere(x):
    return np.sum(np.abs(x),axis=1)
```

```python
options = {'c1': 0.5, 'c2': 0.3, 'w':0.9}
```

```python
# Call instance of PSO
dimensions = 10
min_bound = [-5]*dimensions
max_bound = [5]*dimensions
bounds = (min_bound, max_bound)
optimizer = ps.single.GlobalBestPSO(n_particles=50, dimensions=dimensions, options=options, bounds=bounds)
```

```python
# Perform optimization
cost, pos = optimizer.optimize(funcSphere, iters=500)
```

```python
import matplotlib.pyplot as plt
from pyswarms.utils.plotters import plot_cost_history
plot_cost_history(cost_history=optimizer.cost_history), plt.grid('on')
plt.show()
```
